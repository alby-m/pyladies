```python
import numpy as np
import pandas as pd
```

```python
stolen_cars = pd.read_csv('stolen_cars.csv', sep=';')
print(stolen_cars)
```

```python
print(stolen_cars.info())
```

```python
print(stolen_cars.columns)
```

```python
selection_columns = ['Zjištěno', 'Objasněno- Počet', 'Stíháno; Vyšetřováno Osob - Ženy', 'Stíháno; Vyšetřováno Osob - Celkem', 'Škody V Tis. Kč - Celkem', 'Časová Jednotka']
stolen_cars_selection = stolen_cars[selection_columns].copy()
stolen_cars_selection
```

```python
stolen_cars_split = stolen_cars_selection.join(stolen_cars_selection['Časová Jednotka'].str.split('-', expand = True))
stolen_cars_split.head()
```

```python
stolen_cars_split.columns = ['Auta_Pocet', 'Auta_Objasneno', 'Auta_Stihano_Zen', 'Auta_Stihano_Osob', 'Auta_Skody_Tisice', 'Auta_Casova_Znamka', 'Auta_Mesic', 'Auta_Rok']
stolen_cars_split.head()
```

```python
stolen_bikes = pd.read_csv('stolen_bikes.csv', sep=';')
stolen_bikes.head()
```

```python
stolen_cars_split = stolen_cars_split.rename(columns={'Auta_Casova_Znamka': 'Casova_Znamka'})
stolen_bikes = stolen_bikes.rename(columns={'Kola_Casova_Znamka': 'Casova_Znamka'})
stolen_bikes
```

```python
stolen_cars_per_year = stolen_cars_split.groupby('Auta_Rok').sum()
stolen_cars_per_year
```

```python
stolen_cars_per_year['Auta_Podil_Zeny'] = (100 * stolen_cars_per_year['Auta_Stihano_Zen'])/stolen_cars_per_year['Auta_Stihano_Osob']
stolen_cars_per_year['Auta_Podil_Muzi'] = 100 - stolen_cars_per_year['Auta_Podil_Zeny']
stolen_cars_per_year.head()
```

```python
stolen_cars_per_month = stolen_cars_split.groupby('Auta_Mesic').sum()
stolen_cars_per_month['Auta_Podil_Zeny'] = (100 * stolen_cars_per_month['Auta_Stihano_Zen'])/stolen_cars_per_month['Auta_Stihano_Osob']
stolen_cars_per_month['Auta_Podil_Muzi'] = 100 - stolen_cars_per_month['Auta_Podil_Zeny']
stolen_cars_per_month
```

```python
stolen_bikes_per_month = stolen_bikes.groupby('Kola_Mesic').sum()
stolen_bikes_per_month['Kola_Podil_Zeny'] = (100 * stolen_bikes_per_month['Kola_Stihano_Zen'])/stolen_bikes_per_month['Kola_Stihano_Osob']
stolen_bikes_per_month['Kola_Podil_Muzi'] = 100 - stolen_bikes_per_month['Kola_Podil_Zeny']
stolen_bikes_per_month
```

```python
index_monthly_bikes = stolen_bikes_per_month.reset_index()
sorted_monthly_bikes = index_monthly_bikes.sort_values(axis=0, by='Kola_Mesic')
index_monthly_cars = stolen_cars_per_month.reset_index()
sorted_monthly_cars = index_monthly_cars.sort_values(axis=0, by='Auta_Mesic')
index_yearly_cars = stolen_cars_per_year.reset_index()
sorted_yearly_cars = index_yearly_cars.sort_values(axis=0, by='Auta_Rok')
sorted_monthly_cars
```

```python
sorted_monthly_cars['Auta_Mesic'] = sorted_monthly_cars['Auta_Mesic'].astype('int64')
print(sorted_monthly_cars.dtypes)
```

```python
sorted_monthly_cars = sorted_monthly_cars.sort_values(axis=0, by='Auta_Mesic')
sorted_monthly_cars
```

```python
%matplotlib inline
stolen_cars_split.plot(x='Casova_Znamka', y='Auta_Pocet', figsize=(15,5), grid=True)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
sorted_monthly_cars.plot(x='Auta_Mesic', y='Auta_Pocet', figsize=(15,5), grid=True, ax=ax)
sorted_monthly_bikes.plot(x='Kola_Mesic', y='Kola_Pocet', figsize=(15,5), grid=True, ax=ax)
```
